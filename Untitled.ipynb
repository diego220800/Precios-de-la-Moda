```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_csv("encuestas.csv",delimiter="	",decimal=",")

dfb=np.array(df[df.keys()[62]])
dfm=np.array(df[df.keys()[85]])
precios=np.concatenate((dfb,dfm), axis=0)
print(dfb,dfm, precios)
```

```python
n_b=len(dfb)
n_m=len(dfm)
plt.plot(np.sort(dfb),np.linspace(1/n_b,1,n_b), label="")
plt.plot(np.sort(dfm),np.linspace(1/n_m,1,n_m), label="")
plt.show()
```

```python
real_diff=np.mean(dfb)-np.mean(dfm)
print(real_diff)
```

```python
n_iteraciones= 10000
diffs=np.zeros(n_iteraciones)
for i in range(n_iteraciones):
    np.random.shuffle(precios)
    b=precios[:n_b]
    m=precios[n_b:]
   # print(np.mean(precios), np.mean(b))
    diffs[i]=np.mean(b)-np.mean(m)
```

```python
_ = plt.hist(diffs, bins=40)
plt.vlines(real_diff,0,500,color="red")
plt.show()
```

```python
frac_altos= np.count_nonzero(diffs>real_diff)/len(diffs)
print(frac_altos)
```
